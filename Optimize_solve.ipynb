```python
def subarraySumOptimized(nums, k):
    count = 0
    current_sum = 0
    sum_frequencies = {0: 1}

    for num in nums:
        current_sum += num

        if current_sum - k in sum_frequencies:
            count += sum_frequencies[current_sum - k]

        sum_frequencies[current_sum] = \
            sum_frequencies.get(current_sum, 0) + 1

    return count
print(subarraySumOptimized([3,3,1,-7,3,4], 7)) 
```
