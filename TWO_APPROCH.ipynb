```python
import time
import random

def subarraySumNaive(nums, k):
    count = 0
    n = len(nums)
    for i in range(n):
        s = 0
        for j in range(i, n):
            s += nums[j]
            if s == k:
                count += 1
    return count

def subarraySumOptimized(nums, k):
    count = 0
    curr = 0
    mp = {0: 1}
    for num in nums:
        curr += num
        if curr - k in mp:
            count += mp[curr - k]
        mp[curr] = mp.get(curr, 0) + 1
    return count

sizes = [1000, 10000, 100000]
k = 10

for n in sizes:
    nums = [random.randint(-10, 10) for _ in range(n)]

    start = time.time()
    subarraySumNaive(nums, k)
    naive_time = time.time() - start

    start = time.time()
    subarraySumOptimized(nums, k)
    opt_time = time.time() - start

    print(f"n={n} | Naive={naive_time:.4f}s | Optimized={opt_time:.4f}s")
```

```python
import matplotlib.pyplot as plt

n_values = [1000, 10000, 100000]


naive_times = [0.0549, 4.7871, 441.3312]
optimized_times = [0.0003, 0.0028, 0.0296]

plt.plot(n_values, naive_times, marker='o', label='Naive Algorithm')
plt.plot(n_values, optimized_times, marker='o', label='Optimized Algorithm')

plt.xlabel('Input Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Empirical Analysis of Subarray Sum Equals K')
plt.legend()
plt.grid(True)

plt.show()
```
